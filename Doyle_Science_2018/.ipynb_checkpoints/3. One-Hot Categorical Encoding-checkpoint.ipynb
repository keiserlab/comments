## Doyle Reaction Prediction Pt. 3: One-Hot Categorical Encodings and Out-of-Sample Tests

Fig. 2 from the Doyle paper randomly samples to create a 30% test set. Given that they are randomly sampling from the exhaustive set of all combinations of ligands, substrates, additives, and bases, it's possible that the chemical features aren't necessary to recapitulate Fig. 2.

As in the previous notebook, our hypothesis is:

    Equivalent performance can be achieved without the use of any chemical features.

Design:
    
    We will one-hot encode each feature as its own category, eliminating any notion of chemical information. Each one-hot vector simply provides a reference to the entity, and by definition is incapable of true generalization. We will test this on the equivalent of Fig. 2, as well as the out-of-sample tests as before.

```python
# Import our libraries
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd

# Import relevant scikit-learn modules
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import LinearSVR
from sklearn.metrics import mean_squared_error, r2_score

# Load RPy2 Libraries for R dataframes
from rpy2.robjects import r, pandas2ri
import rpy2.robjects as robjects

import utils

# set a seed for all the randoms
np.random.seed(42)
```

```python
# Define path and constants
DATA_DIR = "data/"
YIELDS_DF = "yields.csv"  # Used modified analysis.R script, see yields.R file
COLS_INFO = "col_layout.csv" # generated from their SI, see above
PLATE_INFO = "plate_layout.csv" # generated from their SI, see above
TRAIN_DF = "training_scaled.rds" # Scaled training data from github
TEST_DF = "test_scaled.rds"      # Scaled test set data from github
```

```python
# Load row and column data
cols = pd.read_csv(DATA_DIR + COLS_INFO)
plates = pd.read_csv(DATA_DIR + PLATE_INFO)
```

```python
# as in the paper, remove controls and additive 7
cols_no_controls   = cols.drop([15, 31, 47]) # remove controls as in paper
plates_no_controls = plates.drop([0, 4, 8, 12, 19, 23, 27, 31]) # remove additive 7 as in paper
```

```python
# Pandas get_dummies one-hot encodes based on the number of variables,
binary_cols = pd.concat([pd.get_dummies(cols_no_controls['Base']), pd.get_dummies(cols_no_controls['Aryl Halide'])], axis=1)
binary_plates = pd.concat([pd.get_dummies(plates_no_controls['Ligand']), pd.get_dummies(plates_no_controls['Additive'])], axis=1)
```

```python
# The dataset is relatively small, just iterate over rows and columns to create each well rxn_fp
vectors = []

for row_idx, row in enumerate(binary_plates.iterrows()):
    row_array = np.array(row[1])
    cols = []
    for col_idx, col in enumerate(binary_cols.iterrows()):
        col_array = np.array(col[1])
        cols.append(np.concatenate([row_array, col_array]))
    vectors.append(cols)
    
vectors = np.array(vectors)
X_set= vectors.reshape(vectors.shape[0] * vectors.shape[1], vectors.shape[2])
```

```python
# Load yields data
yields = pd.read_csv(DATA_DIR + YIELDS_DF)

# As in the paper, remove controls and additive 7

yields = np.array(yields['x'])
yields = yields.reshape(3 * 32, 48)
yields = np.delete(yields, [0, 4, 8, 12, 19, 23, 27, 31], axis=0)
yields = np.delete(yields, [15, 31, 47], axis=1)
yields = yields.flatten()

# it turns out the yield data contains some NaNs
# we set these to zeroes
y_set = yields

for idx, value in enumerate(y_set):
    if np.isfinite(value) == False:
        print(idx, value)
        y_set[idx] = np.nanmean(y_set)
```

```python
# Activate RPy2
pandas2ri.activate()
readRDS = robjects.r['readRDS']

# load R dataframes
train_rds = readRDS(DATA_DIR + TRAIN_DF)
test_rds  = readRDS(DATA_DIR + TEST_DF)

# convert these to pandas dataframes
train_df = pandas2ri.ri2py(train_rds)
test_df  = pandas2ri.ri2py(test_rds)

train_indices = train_df.sort_index().index - 1
test_indices = test_df.sort_index().index - 1
```

```python
X_train = X_set[train_indices]
y_train = y_set[train_indices]

X_test = X_set[test_indices]
y_test = y_set[test_indices]
```

```python
models = [LinearRegression(), # use default
          KNeighborsRegressor(n_neighbors=7), # use k = 7
          LinearSVR(),        # mimic R-default values
          MLPRegressor(hidden_layer_sizes=(5), # mimic the paper values
                       activation='logistic',
                       solver='sgd',
                       max_iter=10000,
                       random_state=42),
          MLPRegressor(hidden_layer_sizes=(100),
                       activation='logistic',
                       solver='sgd',
                       max_iter=10000,
                       random_state=42),
          RandomForestRegressor(n_estimators=500, # 500 trees
                                random_state=42)
         ]
```

```python
preds, r2_values, rmse_values = utils.fit_models(X_train,
                                                           X_test,
                                                           y_train,
                                                           y_test,
                                                           models)

utils.plot_models(preds,
                  r2_values,
                  rmse_values,
                  y_test)
```

If chemical features were critical to this experiment, we would see a considerable reduction in performance. All the plots remain strikingly similar to the prior two notebooks, both with chemical features, and "random" chemical features.

Conclusion:
   
    With random sampling, the random train/test split are inadequate for distinguishing between models trained on chemical features (last notebook) and models trained only on one-hot encoded data.

### Evaluating the Out of Sample Performance

As in the first notebook, only the third plate is included in the paper. We'll use one-hot encoding to do all three plate predictions and compare to notebook 1.

By definition, one-hot encoding can't generalize to new systems, so out-of-sample performance should fail miserably.

```python
# Train on plates 1 and 2
X_train_3 = X_set[:2520]
y_train_3 = y_set[:2520]

# Test on plate 3
X_test_3 = X_set[2520:]
y_test_3 = y_set[2520:]

dataset_3 = [X_train_3, y_train_3, X_test_3, y_test_3]

# Train on plates 1 and 3
X_train_2 = np.concatenate([X_set[:1080], X_set[2520:]])
y_train_2 = np.concatenate([y_set[:1080], y_set[2520:]])

# Test on plate 2
X_test_2 = X_set[1080:2520]
y_test_2 = yields[1080:2520]

dataset_2 = [X_train_2, y_train_2, X_test_2, y_test_2]

# Train on plates 2 and 3
X_train_1 = X_set[1080:]
y_train_1 = y_set[1080:]

# Test on plate 1
X_test_1 = X_set[:1080]
y_test_1 = y_set[:1080]

dataset_1 = [X_train_1, y_train_1, X_test_1, y_test_1]
```

```python
models = [RandomForestRegressor(n_estimators=500,
                                random_state=42),
         RandomForestRegressor(n_estimators=500,
                               random_state=42),
         RandomForestRegressor(n_estimators=500,
                               random_state=42)]
datasets = [dataset_3, dataset_2, dataset_1]

predictions = []
r2_values = []
rmse_values = []

for i, (data, model) in enumerate(zip(datasets, models)):
    model.fit(data[0], data[1])
    preds = model.predict(data[2])
    r_squared = r2_score(data[3], preds)
    rmse = mean_squared_error(preds,  data[3]) ** 0.5
    
    predictions.append(preds)
    r2_values.append(r_squared)
    rmse_values.append(rmse)
    
titles = ['One-Hot Encoded\nRF Plate 3 Predictions',
          'One-Hot Encoded\nRF Plate 2 Predictions',
          'One-Hot Encoded\nRF Plate 1 Predictions']

fig = plt.figure(figsize=(15,5))
colors = ['mediumseagreen','mediumseagreen','mediumseagreen']
for i, color in enumerate(colors):
    fig.add_subplot(1,3,i+1)
    plt.title(titles[i], fontsize=15)
    plt.grid(alpha=0.2)
    r2_patch = mpatches.Patch(label="R2 = {:04.2f}".format(r2_values[i]),color=color)
    rmse_patch = mpatches.Patch(label="RMSE = {:04.1f}".format(rmse_values[i]), color=color)
    plt.xlim(-25,105)
    plt.ylim(0,105)
    plt.scatter(predictions[i], datasets[i][3], alpha=0.2, c=color)
    plt.legend(handles=[r2_patch, rmse_patch], fontsize=12)
    plt.plot(np.arange(100), np.arange(100), ls="--", c=".3")
fig.text(0.5, 0.04, 'predicted', ha='center', va='center', fontsize=15)
fig.text(0.09, 0.5, 'observed', ha='center', va='center', rotation='vertical', fontsize=15)
plt.savefig('plots/one_hot_plates.png', dpi = 300)
plt.show()
```

Plates 2 and 3 are slightly reduced from chemical features, but not by much. Plate 1 somehow exceeds Plate 1 predictions with chemical features (?!). This illustrates the importance of considering multiple folds of a k-fold type experiment.

One-hot encoded features by definition can't generalize, so how are they performing so well here? It appears that each plate prediction has a natural structure to it. The additives do not appear to be that significant based on the performance we see.

Conclusion:
   
    With out-of-sample tests, chemical features minimally outperform one-hot-encoding. Because one-hot categorical encodings cannot generalize, this raises questions about whether the models are learning meaningful chemical features.
